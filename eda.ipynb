```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.rcParams['font.family'] = 'Times New Roman'
```

```python
pos_path = 'data/twitter-datasets/train_pos_full.txt'
neg_path = 'data/twitter-datasets/train_neg_full.txt'

with open(pos_path, 'r') as f:
    pos_tweets = f.readlines()
with open(neg_path, 'r') as f:
    neg_tweets = f.readlines()

print('Number of positive tweets: {}'.format(len(pos_tweets)))
print('Number of negative tweets: {}'.format(len(neg_tweets)))
```

```python
# plot the number of tweets in each class
plt.figure(figsize=(10, 5))
plt.bar(['Positive', 'Negative'], [len(pos_tweets), len(neg_tweets)], color=['blue', 'red'])
plt.xlabel('Class')
plt.ylabel('Number of tweets')
plt.title('Number of tweets in each class')
plt.savefig('./manipulated/num_tweets.png')
plt.show()
```

```python
# plot the length of tweets in each class
plt.figure(figsize=(8, 5))
plt.hist([len(tweet) for tweet in pos_tweets], bins=50, color='blue', alpha=0.5, label='Positive')
plt.hist([len(tweet) for tweet in neg_tweets], bins=50, color='red', alpha=0.5, label='Negative')
plt.xlim(0, 200)
plt.xlabel('Length of tweets')
plt.ylabel('Number of tweets')
plt.title('Length of tweets in each class')
plt.legend()
plt.savefig('./manipulated/length_tweets.png')
plt.show()
```
